#Structure of a basic code snippet in Modal


A code snippet in Modal includes the following components:

•	A Modal App (modal.App)
https://modal.com/docs/reference/modal.App

•	Code Execution Environment (modal.Image):
Defines the runtime environment for the code. Required libraries can be installed here.
https://modal.com/docs/reference/modal.Image

•	Volume (modal.volume)

•	Functions (app.function):
Include the image and volume.

•	Entry Point Function (app.local_entrypoint):
Used to initiate the application locally.


# 6.	Example: Basic CNN Network

To run the code via Modal, we will save our script as a .py file named cnn-dog-cat, and then execute it using:


```
!modal run <file_name.py>

```



````python
#Writing the script

cnn_modal="""
import modal

# Create a Modal app
app = modal.App(name="cnn-dog-cat-classifier")

# Define the Modal image with required dependencies
image = modal.Image.debian_slim(python_version="3.10").pip_install(
    "numpy",
    "matplotlib",
    "tensorflow",
    "pillow"
)

# Connect to the existing volume where the dataset is stored
volume = modal.Volume.from_name("cats_dogs")

# Define the directory path inside the Modal volume
DATA_DIR = "/cats_and_dogs"

@app.function(
    volumes={DATA_DIR: volume},  # Mount the volume inside the function
    image=image,
)
def train_model():
    import os
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
    from tensorflow.keras.preprocessing.image import ImageDataGenerator

    DATA_DIR = "/cats_and_dogs/cats_and_dogs"  # Corrected path
    print(f"Data directory: {DATA_DIR}")

    if not os.path.exists(DATA_DIR):
        raise ValueError(f"Data directory '{DATA_DIR}' does not exist!")

    # Define paths to training and validation directories inside the volume
    train_dir = os.path.join(DATA_DIR, "train")
    val_dir = os.path.join(DATA_DIR, "val")

    # Ensure the dataset exists in the specified location
    if not os.path.exists(train_dir):
        raise ValueError(f"Train directories do not exist in '{DATA_DIR}'")
    if not os.path.exists(val_dir):
       raise ValueError(f"validation directories do not exist in '{DATA_DIR}'")


    # Initialize data generators for loading images
    train_datagen = ImageDataGenerator(rescale=1.0 / 255)
    val_datagen = ImageDataGenerator(rescale=1.0 / 255)

    # Load training images from the volume
    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(150, 150),
        batch_size=32,
        class_mode="binary"
    )

    # Load validation images from the volume
    validation_generator = val_datagen.flow_from_directory(
        val_dir,
        target_size=(150, 150),
        batch_size=32,
        class_mode="binary"
    )

    # Define the CNN model architecture
    model = Sequential([
        Conv2D(32, (3, 3), activation="relu", input_shape=(150, 150, 3)),
        MaxPooling2D(2, 2),
        Conv2D(64, (3, 3), activation="relu"),
        MaxPooling2D(2, 2),
        Conv2D(128, (3, 3), activation="relu"),
        MaxPooling2D(2, 2),
        Flatten(),
        Dense(512, activation="relu"),
        Dropout(0.5),
        Dense(1, activation="sigmoid")
    ])

    # Compile the model
    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

    # Train the model using the dataset from the volume
    history = model.fit(
        train_generator,
        validation_data=validation_generator,
        epochs=5
    )

    # Save the trained model inside the Modal volume
    model_save_path = os.path.join(DATA_DIR, "cat_dog_classifier.keras")
    model.save(model_save_path, save_format="keras")
    print(f"Model trained and saved as '{model_save_path}'")

    # Print final training and validation accuracy
    final_train_acc = history.history["accuracy"][-1]
    final_val_acc = history.history["val_accuracy"][-1]
    print(f"Final Training Accuracy: {final_train_acc:.4f}")
    print(f"Final Validation Accuracy: {final_val_acc:.4f}")

@app.local_entrypoint()
def main():
    print("Starting model training on Modal...")
    train_model.remote()

"""
#Saving the script as a .py file.
with open('/content/cnn-dog-cat.py', 'w') as f:
    f.write(cnn_modal)

#Downloading the .py file.
from google.colab import files
files.download('/content/cnn-dog-cat.py')
````

***Code Explanation:***

**a. Creating an Application in Modal:**
We create an application named "cnn-dog-cat-classifier" in Modal.


```
app = modal.App(name="cnn-dog-cat-classifier")
```



**b. Defining the Image (Execution Environment):**
We define the image, which specifies the environment where the functions will run. Required libraries, such as TensorFlow, NumPy, and Matplotlib, are installed in this step.


```
image = modal.Image.debian_slim(python_version="3.10").pip_install(
    "numpy",
    "matplotlib",
    "tensorflow",
    "pillow"
)
```
**c. Connecting to Storage Using Volume**


```
volume = modal.Volume.from_name("cats_dogs")
```



**d. Defining the Function:**
We define the function using:

```
@app.function(volumes={MODEL_DIR: volume}, image=image)
```


**e. We will write the functions that we want to run in Modal:**

•	Loading the data from the volume.

•	Building the CNN model.

•	Training the model on the data.

•	Saving the trained model back to the volume.

In this case, the function train_model includes all the steps required for training the model.

**f. Running the Function Remotely:**
The function is executed remotely


```
<function name>.remote()
```
